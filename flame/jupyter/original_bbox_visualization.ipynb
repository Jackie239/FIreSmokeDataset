```python
import numpy as np
import matplotlib.pyplot as plt
from bbox import bbox_to_rect
```

```python
imageIndex = 1585

imagePath = '../Images/image_{}.jpg'.format(imageIndex)
txtPath = '../labels/image_{}.txt'.format(imageIndex)

with open(txtPath, 'r') as f:
        txtRead = [x.strip()[1:-1]for x in f.readlines()]
    
bboxes = [x.split() for x in txtRead]
bboxes = np.array(bboxes).astype(np.int64)

imageArray = plt.imread(imagePath)
fig = plt.imshow(imageArray)
for bbox in bboxes:
    fig.axes.add_patch(bbox_to_rect(bbox, 'red'))
```
